```python
def last_page_finder(html):
    page_pkg = html.find_all("table",align="center")
    last_page_pkg = page_pkg[0].find_all("td",class_="pgRR")[0].a.get('href')
    last_page_num = int(last_page_pkg.split("=")[-1].strip())
    return last_page_num
```

```python
def sise_exporter(code) :
    sise_df = pd.DataFrame()
    sise_url = 'https://finance.naver.com/sise/sise_index_day.nhn?code=' + code
    sise_html = BeautifulSoup(urlopen(sise_url).read(), 'lxml') #html.parser
    last_page_num = last_page_finder(sise_html)
    for i in range(1, last_page+1) :
        page_url = (f'{sise_url}&page={i}')
        sise_df = sise_df.append(pd.read_html(page_url, header=0)[0], ignore_index=True)
        sise_df = sise_df.dropna(axis=0)
    return sise_df    
    
```
