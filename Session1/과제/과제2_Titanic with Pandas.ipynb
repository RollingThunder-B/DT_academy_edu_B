# 문제 7 : One Hot Encoding

`Pclass` 컬럼에 대해, `Pclass-1`, `Pclass-2`, `Pclass-3` 컬럼을 만들어주세요.

+ `Pclass-1` : `Pclass=1`이면 `True`, 아니면 `False`

# 문제 13

형제/자매/배우자(SibSp)도 없고, 부모/자식(Parch)도 없는 사람을 `싱글`이라 하겠습니다. `싱글` 컬럼을 만들어주시고 `싱글` 여부에 따른 생존율을 구해주세요.

````python
def single_fn(row) :
    if row['SibSp']==0 and row['Parch']==0 :
        return 'Y'
    return 'N'

train_df['싱글'] = train_df.apply(single_fn, axis=1)
train_df[['Survived', '싱글']].groupby('싱글').mean()
````

# 문제 14

`가족수` 컬럼을 만들어주세요. `가족수 = SibSp + Parch + 1` 수식을 따릅니다.

그리고, `가족수`별 생존율도 구해주세요.

# 문제 15

`가족수` 컬럼이 너무 세분화되어있습니다. 이를 구역으로 나눠보겠습니다.

+ `싱글` : `가족수<=1`
+ `핵가족` : `1<가족수<5`
+ `대가족` : `가족수>=5`

위를 참조해서 `가족형태` 컬럼을 만들어주시고, `가족형태`별 생존율을 구해주세요.

````python
def family_type_fn(row) :
    if row['가족수'] <= 1 : return '싱글'
    elif 1 < row['가족수'] < 5 : return '핵가족'
    return '대가족'

train_df['가족형태'] = train_df.apply(family_type_fn, axis=1)
train_df[['Survived', '가족형태']].groupby('가족형태').mean()
````

# 문제 17

`Name` 컬럼은 `성, 호칭. 이름` 구조를 가지고 있습니다. `호칭`을 추출해서, `호칭` 컬럼에 저장해주세요.

그리고, `호칭`별 생존율을 구해주세요.

````python
def name_title_fn(name):
    return name.split(".")[0].split(",")[1].strip()

train_df['호칭'] = train_df['Name'].apply(name_title_fn)
train_df[['Survived', '호칭']].groupby('호칭').mean()
````

# 문제 18

낮은 모수의 `호칭`들을 정리해봅시다.

`Mr`, `Mrs`, `Miss`, `Master`가 거의 대부분입니다. 이 외의 `호칭`은 `Other`로 정리해봅시다.

그리고, `호칭`별 생존율을 다시 구해봅시다.

````python
def retitle_fn(title) :
    if (title == 'Mr') or (title == 'Mrs') or (title == 'Miss') or (title == 'Master') : return title
    return 'Other'

train_df['호칭'] = train_df['호칭'].apply(retitle_fn)
train_df[['Survived', '호칭']].groupby('호칭').mean()

# title_mask = train_df['호칭'] != ('Mr' or 'Mrs' or 'Master') #안되네
# aaaaa = train_df.loc[title_mask, '호칭'] = ['other']
# aaaaa
````

# 문제 19

`결혼유무` 컬럼을 만들어봅시다.

+ `호칭`이 `Mr`, `Mrs` 이면 : `True`
+ `호칭`이 `Master`, `Miss` 이면 : `False`
+ 나머지 `호칭`은 `NaN` 그대로 둡시다.

`결혼유무`별 생존율을 구해봅시다.

````python
def marry_fn(title) :
    if (title == 'Mr') or (title == 'Mrs') : return True
    elif (title == 'Master') or (title == 'Miss') : return False

train_df['결혼유무'] = train_df['호칭'].apply(marry_fn)
train_df[['Survived', '결혼유무']].groupby(['결혼유무']).mean()
````
