```python
x1 = np.random.rand(5)
x2 = np.random.rand(5)
y = 2*x1 + 3*x2 + 4000
x1,x2,y
```

```python
# shift-M 셀 합치기
# shist-ctrl-'-'  셀 나누기
# a , b  셀 추가
# x 삭제,  o 출력보기

w1 = np.random.rand()
w2 = np.random.rand()
b = np.random.rand()

n_epoch = 200
lr = 1
for epoch in range(n_epoch):
    y_pred = w1*x1 + w2*x2 + b
    e = (y_pred - y)
    w1 = w1 - lr*(e*x1).mean()
    w2 = w2 - lr*(e*x2).mean()
    b = b - lr*e.mean()

    mean_error = np.abs(e).mean()
    print('error', mean_error, 'w1=', w1, 'w2=', w2, 'b=', b)
```

```python
X = np.vstack((x1,x2)).T
X
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X,y)
```

```python
x1=np.random.randn(500)*3 + 170
y1=x1*1.4- 160 + np.random.randn(500)*2

x2=np.random.randn(500)*2.5 + 160
y2=x2*1.1- 120 + np.random.randn(500)*1.5

X = np.hstack((x1,x2))
y = np.hstack((y1,y2))
s = 500*[0]+ 500*[1]
X_all = pd.DataFrame({"height": X, "sex":s, "weight":y})
```
