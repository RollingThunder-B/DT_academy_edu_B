```python
# def age_func(age):
#     if age <= 10 : return 10,
#     elif age <= 20 : return 20,
#     elif age <= 30 : return 30,
#     elif age <= 40 : return 40,
#     elif age <= 50 : return 50,
#     elif age <= 60 : return 60,
#     elif age <= 70 : return 70,
#     elif age <= 40 : return 40,
#     elif age <= 40 : return 40,
#     elif age <= 40 : return 40,
#     else : 200

def age_func(age):
    for i in range(10, 101, 10):
        if age <= i : 
            return (str(i) + ' gen')
    return 'unknown'

train_df['age-label'] = train_df['Age'].apply(age_func)
train_df
```

```python
suv_k = pd.concat([sex_suv_df,
           age_suv_df,
           pclass_suv_df,
           embarked_suv_df]
         , axis=0)
#suv_k = suv_k.to_dict()
suv_k['Survived']
```

```python
import math
```

```python
train_df['k1'] = train_df['Sex'].map(suv_k['Survived'])
train_df['k2'] = train_df['age-label'].map(suv_k['Survived'])
train_df['k3'] = train_df['Pclass'].map(suv_k['Survived'])
train_df['k4'] = train_df['Embarked'].map(suv_k['Survived'])
train_df['kkk'] = np.power(train_df['k1']**2 + train_df['k2']**2 + train_df['k3']**2 + train_df['k4']**2
                          , 1/2)
train_df
```

```python
train_df[['Survived', 'kkk']]
train_df.pivot_table(index='Survived',
                    values='kkk', aggfunc='median')
```

```python
test_df['age-label'] = test_df['Age'].apply(age_func)
test_df
```

```python
test_df['k1'] = test_df['Sex'].map(suv_k['Survived'])
test_df['k2'] = test_df['age-label'].map(suv_k['Survived'])
test_df['k3'] = test_df['Pclass'].map(suv_k['Survived'])
test_df['k4'] = test_df['Embarked'].map(suv_k['Survived'])
test_df['kkk'] = np.power(test_df['k1']**2 + test_df['k2']**2 + test_df['k3']**2 + test_df['k4']**2
                          , 1/2)
test_df
```

```python
def apply_fn(row):
    if row['kkk'] > 1.0 :
        return 1
    return 0

test_df['Survived'] =  test_df.apply(apply_fn, axis=1)
#test_df = test_df.set_index('PassengerId')
```
