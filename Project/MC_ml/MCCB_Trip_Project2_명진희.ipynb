### Trip 시간을  품질 기준으로 상태를 변환 시킴
1. Trip 시간   < 90  : 0
2. 90 <= Trip 시간 <= 150 : 1
3. 150 < Trip  : 2

```python
def trip_fn(row) :
    if row['Trip시간'] < 90 :  # Trip시간 < 90
        return 0
    elif row['Trip시간'] <= 120 :  # Trip시간 <= age <=120
        return 1
    else :
        return 2  # Trip시간 > 120
        
    
mccb_data['trip 상태'] = mccb_data.apply(trip_fn, axis=1)
```

###  데이터 학습을 위해서, 훈련 데이터와 학습 데이터를 나누어 준다.

```python
np.random.seed(10)
X_train, X_test, y_train, y_test = train_test_split(mccb_x, mccb_y , test_size =0.2)
```

### RandomForestRegressor

```python
#하이퍼 파라미터 적용
n = range(10, 50)
for i in n :
    rfr = RandomForestRegressor(n_estimators=i)
    rfr.fit(X_train, y_train)
    print(f"n_estimators {i}의 score : {rfr.score(X_test, y_test)}")
```

```python
rfr = RandomForestRegressor(n_estimators=25)
rfr.fit(X_train, y_train)
y_pred = rfr.predict(X_test)
train_score_rfr =  rfr.score(X_train, y_train)
test_score_rtf = rfr.score(X_test, y_test)
print(f"RandomForestRegressor 훈련  n_estimators= 25의 score : {train_score_rfr}")
print(f"RandomForestRegressor 시험  n_estimators= 25의 score : {test_score_rtf}")
```

```python
pred_data = X_test.copy()

pred_data['y_pred'] = y_pred
pred_data['trip시간'] = y_test
```

```python
pred_data['trip 차이'] = pred_data['trip시간'] - pred_data['y_pred']

pred_data
```

```python
print(f" 실제 Trip 시간 - 예측 Trip 시간의 절대치 최대값 : {abs(pred_data['trip 차이']).max()}")
```

## SVM

 - SVR (Support Vector Regression)은 분류에 SVM과 동일한 원칙을 사용하지만 사소한 차이는 거의 없음

```python
pred_data_svr = X_test.copy()

pred_data_svr['y_pred'] = y_pred_svr
pred_data_svr['trip시간'] = y_test


pred_data_svr['trip 차이'] = pred_data_svr['trip시간'] - pred_data_svr['y_pred']
pred_data_svr
```

```python
print(f" 실제 Trip 시간 - 예측 Trip 시간의 절대치 최대값 : {abs(pred_data_svr['trip 차이']).max()}")
```
