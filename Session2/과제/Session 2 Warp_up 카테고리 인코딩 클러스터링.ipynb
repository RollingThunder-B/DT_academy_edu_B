```python
data_all = pd.read_csv('08860000011_ALL_ResultData_20190702170936.csv')
data_all[:4].T
```

```python
data_all.columns
```

```python
del_list = ['COMP_SEQNO', 'PCB_SEQNO', 'ARRAY_INDEX', 'ASST_SEQNO', 'JOB_NM',
       'GUI_USER', 'USER', 'VENDOR_NM', 'END_DATE_TIME', 
       'COMPONENT_STATUS',
       'Offset Length(um)_RESULT', 'Offset Width (um)_RESULT', 'Offset Angle(Deg)_RESULT',
       'NonBareCopper Area (%)_RESULT', 'Volume (%)_RESULT', 
       'NonBareCopper Area (%)_RESULT.1', 'Volume (%)_RESULT.1',
       'Coner Height [3] (um)_RESULT', 'Coner Height [4] (um)_RESULT', 'Height Diff. 4Corner (um)_RESULT',
       'Max Msr Heiht (um)_RESULT', 'Min Msr Heiht (um)_RESULT',
       'Coner Height [1] (um)_RESULT', 'Coner Height [2] (um)_RESULT', 'Length (%)_VALUE', 'Length (%)_RESULT',
       'Width (um)_VALUE', 'Width (um)_RESULT', 'Thickness (um)_RESULT', 'Width (%)_RESULT',
       'Thickness (%)_RESULT', 'Length (um)_RESULT', 'Volume (%)_RESULT.2',
       'Average Height (um)_RESULT', 'Brightness [Gray]_RESULT',
       'Length Offset Threshold_TOL', 'Width Offset Threshold_TOL',
       'Rotation Offset Threshold_TOL', 'Min. Length[%]_TOL',
       'Max. Length[%]_TOL', 'Max. Width[%]_TOL', 'Min. Width[%]_TOL',
       'Min. Thickness[%]_TOL', 'Max. Thickness[%]_TOL']
data_x = data_all.drop(del_list, axis=1)
data_x[:4].T
```

```python
for i in data_x.columns[:3] :
    print(f'* {i} :')
    print(data_x[i].value_counts())
    print()
    
```

```python
data_x = data_x.drop(['MACHIN_ID', 'INSP_COND_NM'], axis=1, errors='ignore')
data_x[:4].T
```

#### ì¹´í…Œê³ ë¦¬ ì¸ì½”ë”©

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

```python
le = LabelEncoder()
data_x['BARCODE_cate'] = le.fit_transform(data_x['BARCODE'])
data_x[:4].T
```

```python
ohe = OneHotEncoder()
barcode_ohe = ohe.fit_transform(data_x[['BARCODE_cate']]).toarray()
barcode_ohe = pd.DataFrame(barcode_ohe)
data_x = pd.concat([data_x, barcode_ohe], axis=1)
```

```python
data_x = data_x.drop(['BARCODE', 'BARCODE_cate'], axis=1, errors='ignore')
data_x[:5].T
```

```python
list(data_x.columns)
```

```python
sc_list = ['Offset Length(um)_VALUE',
 'Offset Width (um)_VALUE',
 'Offset Angle(Deg)_VALUE',
 'NonBareCopper Area (%)_VALUE',
 'Volume (%)_VALUE',
 'NonBareCopper Area (%)_VALUE.1',
 'Volume (%)_VALUE.1',
 'Coner Height [3] (um)_VALUE',
 'Coner Height [4] (um)_VALUE',
 'Height Diff. 4Corner (um)_VALUE',
 'Max Msr Heiht (um)_VALUE',
 'Min Msr Heiht (um)_VALUE',
 'Coner Height [1] (um)_VALUE',
 'Coner Height [2] (um)_VALUE',
 'Thickness (um)_VALUE',
 'Width (%)_VALUE',
 'Thickness (%)_VALUE',
 'Length (um)_VALUE',
 'Volume (%)_VALUE.2',
 'Average Height (um)_VALUE',
 'Brightness [Gray]_VALUE',]
data_x[sc_list] = sc.fit_transform(data_x[sc_list])
data_x[:5].T
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage, ward
```

```python
Z = linkage(data_x, metric='euclidean', method='ward', ) 

plt.figure(figsize=(12, 6))
plt.title('ë´ë“œë¡œê·¸ëž¨')
dendrogram(Z, labels=data_x.index)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=5)
k = kmeans.fit(data_x)
k
```

```python
my_markers=['*','^', 'o','^','.',',','1','2']
my_color =['r','c','g','b','g','k','r','y']

figure, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2)
figure.set_size_inches(12, 6)

plt.subplot(2, 2, 1)
for n in range(data_x.shape[0]):
    label = k.labels_[n]
    plt.scatter(data_x['Offset Length(um)_VALUE'][n], data_x['Offset Width (um)_VALUE'][n], 
                c=my_color[label], marker=my_markers[label], s=50)
    
plt.subplot(2, 2, 2)
for n in range(data_x.shape[0]):
    label = k.labels_[n]
    plt.scatter(data_x['Offset Length(um)_VALUE'][n], data_x['Offset Angle(Deg)_VALUE'][n], 
                c=my_color[label], marker=my_markers[label])
    
plt.subplot(2, 2, 3)
for n in range(data_x.shape[0]):
    label = k.labels_[n]
    plt.scatter(data_x['Volume (%)_VALUE'][n], data_x['Coner Height [3] (um)_VALUE'][n], 
                c=my_color[label], marker=my_markers[label])
    
plt.subplot(2, 2, 4)
for n in range(data_x.shape[0]):
    label = k.labels_[n]
    plt.scatter(data_x['Coner Height [3] (um)_VALUE'][n], data_x['Height Diff. 4Corner (um)_VALUE'][n], 
                c=my_color[label], marker=my_markers[label])
```

```python
data_clus = data_x
data_clus['k_labels'] = k.labels_
data_clus[:3]
```

```python
figure, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2)
figure.set_size_inches(12, 8)

sns.scatterplot(data=data_x, x="Offset Length(um)_VALUE", y="Offset Width (um)_VALUE" , hue='k_labels', palette = "Set1", ax=ax1)
sns.scatterplot(data=data_x, x="Offset Length(um)_VALUE", y="Offset Angle(Deg)_VALUE", hue='k_labels', palette = "Set1", ax=ax2)
sns.scatterplot(data=data_x, x="Volume (%)_VALUE", y="Coner Height [3] (um)_VALUE", hue='k_labels', palette = "Set1", ax=ax3)
sns.scatterplot(data=data_x, x="Coner Height [3] (um)_VALUE", y="Height Diff. 4Corner (um)_VALUE", hue='k_labels', palette = "Set1", ax=ax4)
```
